```python
import Q_Value_Agent1 as trainingAgent
import ice_env as environment
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt 
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
env = environment.env()
```

```python
agent1 = trainingAgent.agent()
agent1.run(alpha=1.0, gamma=1.0, numGames=10000, env=env, model = 'max_qValue_policy')
```

```python
agent1.episode_df.to_csv('episodes.csv')
```

```python
df_episode_hour = agent1.episode_df
sum_switch=df_episode_hour[['Episode','Switch']].groupby(['Episode']).sum().reset_index()
df_episode = df_episode_hour[df_episode_hour['Current_Hour']==23].drop(['New_Ice_Needed','Current_Hour','Action','Reward','Q_val','Switch'],axis=1).reset_index(drop=True)
df_episode = pd.merge(df_episode,sum_switch,how='left',on=['Episode'])
df_episode['Ice_Wasted'] = df_episode['Diff']+df_episode['Ice_Melted']
#df_episode['Wasted_Rate'] = df_episode['Ice_Wasted']/df_episode['Ice_Needed']
df_episode
```

```python
average_wasted = sum(list(df_episode['Ice_Wasted']))/len(list(df_episode['Ice_Wasted']))
average_needed = sum(list(df_episode['Ice_Needed']))/len(list(df_episode['Ice_Needed']))

average_melted = sum(list(df_episode['Ice_Melted']))/len(list(df_episode['Ice_Melted']))
average_diff = sum(list(df_episode['Diff']))/len(list(df_episode['Diff']))
```
